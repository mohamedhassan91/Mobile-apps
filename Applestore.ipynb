```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

import seaborn as sns
```

```python
dataset= pd.read_csv('AppleStore.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
def paid (x):
    if x > 0:
        return 'paid'
    else:
        return 'free'
dataset['category'] = dataset.price.apply(lambda x : paid(x))
dataset.head()
```

```python
dataset['user_rating_ver'].hist()
```

```python
dataset.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
f, ax = plt.subplots(figsize=(16, 10))
sns.countplot(y="prime_genre", data=dataset, color="c")
```

```python
dataset['user_rating_ver'].value_counts()
```

```python
dataset.groupby(by='prime_genre').sum()
```

```python
dataset.prime_genre.value_counts().plot(kind='bar')
```

```python
dataset.category.value_counts().plot(kind='pie')
```

```python
sns.catplot(data=dataset, kind="bar", x="user_rating", y="prime_genre", hue="category")
```

```python
dataset.category.value_counts().plot(kind='bar')
```
